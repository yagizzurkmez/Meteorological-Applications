```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Generate a random data from uniform distribution
data = np.random.uniform(1, 6, 100) # Dice example, for 100 times pick
mean= data.mean()
sigma = data.std()
print("Mean: ", mean)
print("Standard Deviation: ", sigma)


#Plot the pdf of the data
sns.histplot(data, kde=True, stat='density') 
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1, label=f'Mean: {mean:.2f}')
plt.axvline(mean - sigma, color='orange', linestyle='dashed', linewidth=1, label=f'Mean - 1σ: {mean - sigma:.2f}')
plt.axvline(mean + sigma, color='orange', linestyle='dashed', linewidth=1, label=f'Mean + 1σ: {mean + sigma:.2f}')
plt.legend()
plt.show()



```

```python
#If we repeat the generate randomly pick 100 from Uni. Dist. more than 1 and find how mean and sigma distrubitons change

iterations = [10, 100, 1000]  # Updated iterations

for iter in iterations:
    means = []
    sigmas = []
    for _ in range(iter):
        data = np.random.uniform(1, 6, 100)
        means.append(data.mean())
        sigmas.append(data.std())
    
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    
    # Plot the distribution of means
    sns.histplot(means, kde=True, ax=axs[0])
    axs[0].set_title(f'Means distribution for {iter} iterations')
    
    # Plot the distribution of sigmas
    sns.histplot(sigmas, kde=True, ax=axs[1])
    axs[1].set_title(f'Sigmas distribution for {iter} iterations')
    
    plt.tight_layout()
    plt.show()
```

```python
# So if we pick randomly 100 number from uniform distribution
# And we repeat this process more than 1 time
# We can see that the means and sigmas are distributed normally if we increase the number of iterations
```
